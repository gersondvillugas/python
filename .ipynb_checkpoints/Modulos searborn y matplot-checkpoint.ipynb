```python
%matplotlib inline
```

```python
propinas=ans.load_dataset('tips')
propinas.head(10)
```

```python
 ans.lmplot('total_bill','tip',propinas)
```

```python
ans.lmplot('total_bill','tip',propinas,scatter_kws={'marker':'0','color':'green'},line_kws={'color':'blue'})
```

```python
ans.lmplot('total_bill','tip',propinas,fit_reg=False)
```

```python
propinas.head()
```

```python
propinas['porcentaje']=100*propinas['tip']/propinas['total_bill']
propinas.head()
```

```python
ans.lmplot('size','porcentaje',propinas)
```

```python
ans.lmplot('total_bill','porcentaje',propinas,hue='sex',markers=['x','o'])
```

```python
ans.lmplot('total_bill','porcentaje',propinas,hue='day')
```

```python
#Mapas de calor (heatmap)
%matplotlib  inline
```

```python
vuelos=ans.load_dataset('flights')
vuelos.head()
```

```python
vuelos=vuelos.pivot('month','year','passengers')
vuelos
```

```python
ans.heatmap(vuelos)
```

```python
ans.heatmap(vuelos,annot=True,fmt='d')
```

```python
valor_central=vuelos.loc['May'][1956]
```

```python
valor_central
```

```python
ans.heatmap(vuelos,center=valor_central,annot=True,fmt='d')
```

```python
ans.heatmap(vuelos,center=valor_central,annot=True,fmt='d',cbar_kws={'orientation':'horizontal'})
```
